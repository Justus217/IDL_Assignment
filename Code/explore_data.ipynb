```python
import torch
```

```python
cells_path = "../data/cells.pt" 
chest_path = "../data/chest.pt"
lesion_path = "../data/lesions.pt"
organs_path = "../data/organs.pt"
orgs_path = "../data/orgs.pt"

cells = torch.load(cells_path)
chest = torch.load(chest_path)
lesion = torch.load(lesion_path)
organs = torch.load(organs_path)
orgs = torch.load(orgs_path)

datasets = {"cells" : cells, "chest" : chest, "lesion" : lesion, "organs" : organs, "orgs" : orgs}
```

```python
for name, dataset in datasets.items():
    print(f"Dataset: {name}")
    print(f"Number of samples: {len(dataset)}")
    print(f"Train sample shape: {dataset['train_images'].shape}")
    print(f"Train sample shape: {dataset['train_labels'].shape}")
    print(f"Test sample shape: {dataset['test_images'].shape}")
    print(f"Test sample shape: {dataset['test_labels'].shape}")
```

```python
for name, data in datasets.items():
    unique_labels = torch.unique(data['train_labels'])
    # Anzahl der Klassen berechnen  
    num_classes = unique_labels.numel()
    print(f"Anzahl der Klassen im datensatz {name} (num_classes): {num_classes}")
```
